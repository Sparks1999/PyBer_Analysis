# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
rides_per_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_per_city_type
```

```python
# 2. Get the total drivers for each city type

drivers_per_city_type = city_data_df.groupby("type").sum()["driver_count"]
drivers_per_city_type
```

```python
#  3. Get the total amount of fares for each city type
fares_per_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_per_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = fares_per_city_type / rides_per_city_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = fares_per_city_type / drivers_per_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
ride_share_summary = pd.DataFrame()
    
ride_share_summary["Total Rides"] = rides_per_city_type
ride_share_summary["Total Drivers"] = drivers_per_city_type
ride_share_summary["Total Fares"] = fares_per_city_type
ride_share_summary["Average Fare Per Ride"] = average_fare_per_ride
ride_share_summary["Average Fare Per Driver"] = average_fare_per_driver

ride_share_summary
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
ride_share_summary.index.name = None
```

```python
#  8. Format the columns.
ride_share_summary["Average Fare Per Driver"].map("${:.2f}".format)
ride_share_summary["Total Rides"] = ride_share_summary["Total Rides"].map("{:,}".format)
ride_share_summary["Total Drivers"] = ride_share_summary["Total Drivers"].map("{:,}".format)
ride_share_summary["Total Fares"] = ride_share_summary["Total Fares"].map("${:,.2f}".format)
ride_share_summary["Average Fare Per Ride"] = ride_share_summary["Average Fare Per Ride"].map("${:.2f}".format)
ride_share_summary["Average Fare Per Driver"] = ride_share_summary["Average Fare Per Driver"].map("${:.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_city_merge = pyber_data_df.groupby(["type", "date"]).sum().fare.to_frame()
fare_city_merge
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_city_merge = fare_city_merge.reset_index()
fare_city_merge
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_city_merge = fare_city_merge.pivot(index="date", columns="type", values="fare")
fare_city_merge
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
fare_and_date = fare_city_merge.loc['2019-01-01':'2019-04-29']
fare_and_date
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
type(fare_and_date.index)
fare_and_date.index = pd.to_datetime(fare_and_date.index)

fare_and_date
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fare_and_date.dtypes
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_and_date = fare_and_date.resample('W').sum()
fare_and_date
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fare_and_date.plot(figsize=(18, 5))
plt.ylabel('Fare ($USD)')
plt.title('Total Fare by City')
plt.savefig('analysis/pyber_summary.png');
```
